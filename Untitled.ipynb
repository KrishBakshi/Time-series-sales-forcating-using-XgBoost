```python
def create_first_last_5_days_features(df, date_column):
    """
    Adds two features to the DataFrame:
    - 'first_5_days_of_month': True for the first 5 days of the month.
    - 'last_5_days_of_month': True for the last 5 days of the month.
    
    Parameters:
    df (pd.DataFrame): DataFrame containing a date column.
    date_column (str): Name of the column containing date values.
    
    Returns:
    pd.DataFrame: DataFrame with two new boolean columns.
    """
    # Ensure the date column is in datetime format
    df[date_column] = pd.to_datetime(df[date_column])
    
    # Extract the day of the month
    df['day'] = df[date_column].dt.day
    
    # Create 'first_5_days_of_month' feature
    df['first_5_days_of_month'] = df['day'] <= 5
    
    # Create 'last_5_days_of_month' feature
    df['last_5_days_of_month'] = df['day'] >= (df[date_column].dt.days_in_month - 4)
    
    # Drop the 'day' column if not needed
    df.drop(columns='day', inplace=True)
    
    return df
```

```python
def handle_outliers(train, column, method='remove'):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = train[column].quantile(0.05)
    Q3 = train[column].quantile(0.95)
    IQR = Q3 - Q1

    # Define the bounds for outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    if method == 'remove':
        # Remove outliers
        train = train[(train[column] >= lower_bound) & (train[column] <= upper_bound)]
    elif method == 'cap':
        # Cap outliers
        train[column] = train[column].apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

    return train
```

```python
## store average sales
def add_store_daily_average_sales(df):
   
    
    # Step 1: Group by STORE_ID and CALDAY, sum SALE_QTY to get total sales per day per store
    store_daily_sales = df.groupby(['STORE_ID', 'CALDAY'])['TOTAL_BASE_SALE_QTY'].sum().reset_index()
    
    # Step 2: Compute the daily average sales for each store
    store_daily_avg_sales = store_daily_sales.groupby('STORE_ID')['TOTAL_BASE_SALE_QTY'].mean().reset_index()
    
    # Rename the columns
    store_daily_avg_sales.columns = ['STORE_ID', 'store_daily_average_sales']
    
    # Step 3: Merge the 'store_daily_average_sales' back into the original dataframe
    df = pd.merge(df, store_daily_avg_sales, on='STORE_ID', how='left')
    
    return df
```

```python
combined_df = add_store_daily_average_sales(combined_df)
```

```python
def add_store_tier(df):
    
    
    # Check if 'store_daily_average_sales' exists in the dataframe
    if 'store_daily_average_sales' not in df.columns:
        raise ValueError("The dataframe must contain the 'store_daily_average_sales' column.")
    
    # Step 1: Create quantiles for the tiers
    # 3: Top 25%, 2: 50-75%, 1: 25-50%, 0: Bottom 25%
    df['store_tier'] = pd.qcut(df['store_daily_average_sales'], q=4, labels=[0, 1, 2, 3], duplicates='drop')
    
    return df
```

```python
combined_df = add_store_tier(combined_df)
```

```python
def add_product_tier(df):
    """
    Adds a 'product_tier' feature to the dataframe based on the distribution of 
    total sales ('SALE_QTY') for each product. Products are divided into 4 tiers:
    - 3: Top 25% of products with the highest total sales
    - 2: Upper-middle 25% of products
    - 1: Lower-middle 25% of products
    - 0: Bottom 25% of products with the lowest total sales

    Parameters:
    df (pd.DataFrame): The input dataframe containing 'PRODUCT_ID' and 'SALE_QTY' columns.

    Returns:
    pd.DataFrame: The original dataframe with an additional 'product_tier' column.
    """
    
    # Step 1: Group by PRODUCT_ID and sum SALE_QTY to get total sales per product
    product_total_sales = df.groupby('PRODUCT_ID')['TOTAL_BASE_SALE_QTY'].sum().reset_index()
    
    # Step 2: Create quantiles for the product tiers
    product_total_sales['product_tier'] = pd.qcut(product_total_sales['TOTAL_BASE_SALE_QTY'], q=4, labels=[0, 1, 2, 3], duplicates='drop')
    
    # Step 3: Merge the 'product_tier' back into the original dataframe
    df = pd.merge(df, product_total_sales[['PRODUCT_ID', 'product_tier']], on='PRODUCT_ID', how='left')
    
    return df
```

```python
combined_df = add_product_tier(combined_df)
```

```python
def one_hot_encode(df, columns):
    """
    Perform one-hot encoding on specified columns in a DataFrame.
    
    Parameters:
    df (pd.DataFrame): The input DataFrame.
    columns (list): List of columns to be one-hot encoded.
    
    Returns:
    pd.DataFrame: The DataFrame with one-hot encoded columns.
    """
    # Perform one-hot encoding using pd.get_dummies
    df_encoded = pd.get_dummies(df, columns=columns, drop_first=True)
    
    return df_encoded
    
```

```python
def MAPE_EVAL(df_max, col):
    APE = []
    for index, row in df_max.iterrows():
        # Handle case where actual sales (SALE_QTY) is zero
        actual_sales = row['TOTAL_BASE_SALE_QTY'] if row['TOTAL_BASE_SALE_QTY'] != 0 else 1
        
        # Calculate percentage error without rounding off
        if row[col] is not None:
            per_err = (row['TOTAL_BASE_SALE_QTY'] - row[col]) / actual_sales
        else:
            per_err = (row[col] - row['TOTAL_BASE_SALE_QTY'])
        
        per_err = np.abs(per_err)
        APE.append(per_err)

    MAPE = np.sum(APE) / len(APE)

    # Print the MAPE value and percentage
    print(f'''
    MAPE   : {MAPE}
    MAPE % : {MAPE * 100} %
    ''')
```
